```python
import pandas as pd
import matplotlib.pyplot as plt

pruebas_miso = pd.read_excel("Estrategia de Pruebas MISO (respuestas)(1).xlsx")
pruebas_miso_clean = pruebas_miso[['NOMBRE_EMPRESA','AREA_TRABAJO',
                                   'POSICION', 'AÑOS_EXPERIENCIA', 
                                   'PREGUNTA_1', 'PREGUNTA_3']]
```

# Solo PREGUNTA_1

```python
pruebas_miso_clean[["PREGUNTA_1"]].groupby("PREGUNTA_1").size().to_frame('size')
```

# solo PREGUNTA_3

```python
pruebas_miso_clean[["PREGUNTA_3"]].groupby("PREGUNTA_3").size().to_frame('size')
```

# Solo AÑOS_EXPERIENCIA

```python
plt.hist(pruebas_miso_clean['AÑOS_EXPERIENCIA'])
plt.show()
```
